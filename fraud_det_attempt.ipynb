```python
import pandas as pd
```

Load only first 21 rows

```python
df = pd.read_csv('transactions_data.csv', nrows=21)
```

Define weights

```python
weights = {
    'amount': 0.3,
    'use_chip': 0.3,
    'date': 0.2,
    'merchant_city': 0.2
}
```

Convert 'date' column to datetime

```python
df['date'] = pd.to_datetime(df['date'])
```

Extract hour from time

```python
df['hour'] = df['date'].dt.hour
```

Clean and convert 'amount'

```python
df['amount'] = df['amount'].astype(str).str.replace(r'[\$,]', '', regex=True).str.strip()
df['amount'] = pd.to_numeric(df['amount'], errors='coerce')
```

âœ… Amount risk using np.where or apply

```python
df['amount_risk'] = df['amount'].apply(lambda x: 1 if x > 50 else 0.5) * weights['amount']
```

Time risk

```python
df['timeRisk'] = df['hour'] * weights['date']
```

use_chip risk

```python
df['chipUseRisk'] = df['use_chip'].apply(lambda x: 1 if x == 'Swipe Transaction' else 0.5) * weights['use_chip']
```

location risk

```python
df['locationRisk'] = df['merchant_city'].apply(lambda x: 1 if x == 'Beulah' else 0.5) * weights['merchant_city']
```

Total risk

```python
df['totalRisk'] = df['amount_risk'] + df['chipUseRisk'] + df['locationRisk'] + df['timeRisk']
```

identify risk level

```python
df['risk_flag'] = df['totalRisk'].apply(lambda x: 'High Risk' if x > 0.6 else 'Low/Moderate')
```

Output relevant columns

```python
print(df[['amount', 'amount_risk', 'chipUseRisk', 'locationRisk', 'timeRisk', 'totalRisk', 'risk_flag']])
```
